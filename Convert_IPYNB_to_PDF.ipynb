```python
import json
import os
from google.colab import files
from nbconvert import PDFExporter
import nbformat

def ubah_metadata_dari_nama_file(nama_file):
    """Mengubah metadata file IPYNB dengan judul diambil dari nama file."""
    try:
        nama_file_tanpa_ext = os.path.splitext(nama_file)[0]
        judul = nama_file_tanpa_ext.replace("_", " ")
        judul = judul.title()

        with open(nama_file, 'r') as f:
            data = json.load(f)

        if 'metadata' not in data:
            data['metadata'] = {}

        data['metadata']['title'] = judul

        with open(nama_file, 'w') as f:
            json.dump(data, f, indent=2)

        print(f"Metadata 'title' berhasil ditambahkan dengan nilai '{judul}'")
        print(f"File '{nama_file}' telah ditimpa.")
        return judul # mengembalikan judul untuk digunakan saat konversi ke pdf

    except FileNotFoundError:
        print(f"File '{nama_file}' tidak ditemukan. Pastikan file telah diupload.")
        return None  # mengembalikan None jika file tidak ditemukan
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
        return None

def konversi_ke_pdf(nama_file, judul_pdf):
    """Konversi file IPYNB ke PDF."""
    try:
        with open(nama_file, 'r') as f:
            notebook_content = f.read()

        notebook_node = nbformat.reads(notebook_content, as_version=4)

        pdf_exporter = PDFExporter()
        pdf_exporter.exclude_input = False
        pdf_exporter.template_name = "lab"
        pdf_exporter.extra_template_data = {'title': judul_pdf}  # Gunakan judul yang telah diupdate

        (pdf_data, resources) = pdf_exporter.from_notebook_node(notebook_node)

        pdf_path = nama_file[:-6] + ".pdf"
        with open(pdf_path, 'wb') as pdf_file:
            pdf_file.write(pdf_data)

        print(f"PDF successfully generated: {pdf_path}")
        files.download(pdf_path)
        print(f"PDF downloaded to your local machine.")

    except Exception as e:
        raise RuntimeError(f"Failed to convert notebook to PDF: {e}")


# Instruksi upload
print("Silakan upload file IPYNB Anda melalui 'Choose Files' di panel sebelah kiri.")

uploaded = files.upload()

for nama_file, _ in uploaded.items():
    judul = ubah_metadata_dari_nama_file(nama_file)

    if judul: # Jika metadata berhasil diupdate
        konversi_ke_pdf(nama_file, judul)
    else:
        print("Proses konversi ke PDF dibatalkan karena gagal memperbarui metadata.")
```
